<h1 align="center">Задание 1</h1>

Большинству из вас знакома форма записи на бесплатные занятия Нетологии.
Предлагаем именно на ней попробовать себя в роли тестировщика и провести тестирование позитивных и негативных сценариев.

Чтобы найти действующую форму записи на занятие, перейдите по ссылке и кликните на заголовок любого из предстоящих бесплатных занятий. Вы попадёте на лендинг о занятии с формой записи:

Начнём с позитивной проверки. Как правило, позитивных проверок всегда намного меньше, чем негативных. Основная задача позитивных проверок — протестировать систему на выполнение ключевых задач при вводе валидных данных. Вариантов, когда «что-то пошло не так» всегда намного больше, чем успешных выполнений операции.

Создайте гугл-документ, откройте доступ к просмотру.
Напишите в нём 2–3 способа позитивной проверки формы записи на занятие.
Ссылку на документ прикрепите в окно «Решение» и нажмите кнопку «Отправить». В ответ вы получите наши способы позитивных проверок.

<h1 align="center">Позитивное тестирование</h1>

Прошло успешно:

- Запись на занятие
- Ввод в поле корректного номера телефона
- Верный переход по ссылкам политики и пользовательского соглашения


<h1 align="center">Задание 2</h1>

Переходим к негативным проверкам. Обычно вариантов ввести данные неправильно намного больше, чем правильно. Напишите 3–5 способов негативной проверки формы записи на занятие.

<h1 align="center">Негативное тестирование</h1>

Безуспешно:
    
- Попытка записи без ввода имени
- Попытка записи без ввода номера телефона
- Попытка записи с вводом некорректного формата телефона
