```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SpaceX_Launches_Data.csv')
df.info()
```

<h3 align="center">Столбцы</h3>

### Количественные:
- PayloadMass - масса полезной нагрузки

### Категориальные:
- Longitude - долгота
- Latitude - широта
- FlightNumber - номер запуска
- Date - дата
- BoosterVersion - версия ракеты
- Orbit - орбита
- LaunchSite - стартовая площадка
- Outcome - результат
- Flights - полёт
- GridFins - решётчатые стабилизаторы
- Reused - повторное использование
- Legs - выдвижные опоры
- LandingPad - посадочная площадка
- Block - компоновка
- ReusedCount - количество раз повторного использования
- Serial - серия

```python
df
```

### Гипотеза

- Значения NaN в столбце Outcome - это миссии не предполагающие возврата 1-ой ступени, так как в этих миссиях отсутствуют выдвижные опоры, то есть речь в датафрейме идёт о возвращаемых ступенях.

```python
df_legs = df.loc[df['Legs'].isin([False])]
```

```python
plt.pie(df_legs.groupby('Legs')['Outcome'].count());
```

- Большинство миссий с выдвижными опорами оказались успешными

```python
df_outcome = df.loc[df['Outcome'].isin(['True ASDS','True RTLS','True Ocean'])]
```

```python
plt.pie(df_outcome.groupby('Legs')['Outcome'].count());
```

- Большинство многоразовых полётов 1-ой ступени связаны с реализацией конструктивно-компоновочной схемы Block 5.0

```python
sns.countplot(x='Serial', hue='Block', data=df);
```

- Cущественный рост выводимой полезной нагрузки при реализации многоразовой ракеты версии Block 5.0

```python
sns.boxplot(x='Block', y='PayloadMass', data=df);
```

- матрица корреляций сообщает, что координаты стартовых площадок SpaceX практически не повлияли на рост полезной нагрузки

```python
sns.heatmap(df.corr(), cmap='Spectral', annot=True, vmin=-1, vmax=1);
```

- матрица диаграмм подтверждает сделанные ранее выводы

```python
sns.pairplot(df[['PayloadMass', 'Block', 'ReusedCount']]);
```
