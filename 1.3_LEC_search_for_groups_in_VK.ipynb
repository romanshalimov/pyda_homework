Напишем программу для проверки логина и пароля

1) логин и пароль пользователя хранится в переменной

2) будем давать пользователю k попыток на ввод логина и пароля

3) если после трех попыток вверного ввода не было, то заблокируем вход - вывести на экран сообщение "Вход заблокирован"

4) если пароль и логин были введены верно, то вывести на экран сообщение "Добро пожаловать!"

```python
login_correct = 'admin'
pswd_correct = 'admin'
k = 3
for i in range(0, k):
    login = input('Ввелите логин')
    pswd = input('Введите пароль')
    if login != login_correct or pswd != pswd_correct:
        k -= 1
        print(f'Осталось {k} попытки')
    else:
        print('Добро пожаловать!')
        break
else:
    print('Вход заблокирован')
```

Посчитаем сколько раз каждое слово встречалось в тексте и выведем информацию в виде:  
`"слово": "количество вхождений" `


```python
some_text = '''
Python является мультипарадигмальным языком программирования, поддерживающим императивное, 
процедурное, структурное, объектно-ориентированное программирование, метапрограммирование и функциональное программирование.
'''
```

```python
import re
res = {}

splitted_text = re.sub(r'[^\w\s]', '', some_text.lower()).split()

for word in splitted_text:
    # print(word)
    if word not in res:
        res[word] = 1
    else:
        res[word] += 1

res
```

```python
res = {}

splitted_text = re.sub(r'[^\w\s]', '', some_text.lower()).split()

for word in splitted_text:
    # print(word)
    res.setdefault(word, 0)
    res[word] += 1

for word, count in res.items():
    print(f'{word}: {count}')
```

```python
splitted_text = re.sub(r'[^\w\s]', '', some_text.lower()).split()
# splitted_text = []
if splitted_text:
    for word in set(splitted_text): 
        print(f'{word}: {splitted_text.count(word)}')
else:
    print('Пустой текст')
```

Посчитаем количество лайков в сообщениях

```python
params = {
    'access_token': token,
    'v':'5.131',
    'q': 'Нетология',
    'start_from': '',
    'count': 200,
    'extended': 1
}   
res = []
while True:
    result = requests.get('https://api.vk.com/method/newsfeed.search', params)
    time.sleep(0.33)
    res += result.json()['response']['items']
    if 'next_from' in result.json()['response'].keys():
        params['start_from'] = result.json()['response']['next_from']
    else:
        break
```

```python
print(type(res))
print(len(res))
```

```python
res[0]
```

```python
# посчитаем количество лайклов и репостов
likes_count = 0
reposts_count = 0
for post in res:
    likes_count += post['likes']['count']
    reposts_count += post['reposts']['count']

print(likes_count)
print(reposts_count)
```

```python
# отфильтруем сообщение, где просмотров было больше 1000

# [post for post in res if post['views']['count'] > 1000]
[post for post in res if 'views' in post and 'count' in post['views'] and post['views']['count'] > 1000]
```
