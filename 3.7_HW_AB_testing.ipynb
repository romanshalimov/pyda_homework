##### Задание № 1
Вам даны результаты проверки двух препаратов.
    
Требуется:
- Оценить, есть ли различие в эффекте у этих пациентов
- Визуально подтвердить результат
    
`A` - результаты для группы людей получающих препарат А.

`B` - результаты для другой группы людей, получающих препарат B.

```python
np.random.seed(11)
A = stats.norm.rvs(scale=30, loc=11, size=100)
B = A + stats.norm.rvs(scale=54, loc=11, size=100)
```

Так как выборки независимые (люди разные) и распределение нормальное, применяем t-тест Стьюдента для независимых выборок
- Н0 - различие в препаратах есть

```python
t, p_value = stats.ttest_ind(A, B)

if p_value < 0.05:
    print('Отклонить Н0')
else:
    print('Не отклонять Н0, различия в препаратах нет')
```

```python
print(t.round(3))
print(p_value.round(3))
```

```python
plt.hist(A, bins=20);
plt.hist(B, bins=20, alpha=0.5);
```

Действительно ли сумма характеристик `HP`, `Attack`, `Defense` у легендарных покемонов выше, чем у других покемонов? А произведение этих же параметров? 

Найдите ответы на эти вопросы и убедите всех в своем выводе статистически.

*Вы можете предположить, что распределение сум и произведений этих параметров является нормальным для всех классов покемонов.*

```python
df_pokemon = pokemon[(pokemon['Legendary'] == True)][['HP','Attack','Defense']]
df_pokemon['Sum'] = df_pokemon['HP'] + df_pokemon['Attack'] + df_pokemon['Defense']
df_pokemon['Prod'] = df_pokemon['HP'] * df_pokemon['Attack'] * df_pokemon['Defense']
df_pokemon.head()
```

```python
df2_pokemon = pokemon[(pokemon['Legendary'] == False)][['HP','Attack','Defense']]
df2_pokemon['Sum'] = df2_pokemon['HP'] + df2_pokemon['Attack'] + df2_pokemon['Defense']
df2_pokemon['Prod'] = df2_pokemon['HP'] * df2_pokemon['Attack'] * df2_pokemon['Defense']
df2_pokemon.head()
```

H0 - средняя сумма показателей у поколений покемонов одинаковая

```python
result = stats.ttest_ind(df2_pokemon['Sum'], df_pokemon['Sum'], equal_var=False)
a = 0.05

if (result.pvalue < a):
    print('Отклонить Н0, средняя сумма показателей у поколений покемонов разная')
else:
    print('Не отклонять Н0, средняя сумма показателей у поколений покемонов одинаковая')
```

Средняя сумма показателей у легендарных покемонов больше, чем у обычных.

H0 - среднее произведение у поколений покемонов одинаковое

```python
result = stats.ttest_ind(df2_pokemon['Prod'], df_pokemon['Prod'], equal_var=False)
a = 0.05

if (result.pvalue < a):
    print('Отклонить Н0, средняя сумма показателей у поколений покемонов разная')
else:
    print('Не отклонять Н0, средняя сумма показателей у поколений покемонов одинаковая')
```

Среднее произведение показателей у легендарных покемонов больше, чем у обычных.
