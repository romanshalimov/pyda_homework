## 3. Functions and working with data

### Task 1
Write a function that classifies films from the lesson materials according to the following rules:

- score 2 and less - low rating
- score 4 and less - average rating
- score 4.5 and 5 - high rating

Write the classification result in the 'class' column.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')

df = movies.merge(ratings, on = 'movieId', how = 'left')
df_groupby_object = df.groupby(['title'])
films_classification = df_groupby_object.rating.mean().reset_index()

def rating_classification(row):
    
    if row['rating'] <= 2:
        return 'Low'
    elif row['rating'] <= 4:
        return 'Average'
    else:
        return 'High'

films_classification['class'] = films_classification.apply(rating_classification, axis = 1)

films_classification.head(10)
```

### Task 2

We use the keywords.csv file.

It is necessary to write a geo-classifier that will be able to set the geographical affiliation of a certain region to each row. I.e., if the search query contains the name of the city of the region, then the name of this region is written in the ‘region’ column. If the search query does not contain the name of the city, then put ‘undefined'.

Rules of distribution by region Center, North-West and Far East:

Write the classification result in a separate column region.

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

df = pd.read_csv('keywords.csv')

def regions(row):
    for region, cities in geo_data.items():
        for city in cities:
            if city in row['keyword']:
                return region
    return 'undefined'

df['region'] = df.apply(regions, axis=1)

show = df[df['region'] != 'undefined']
print(show[0:10])
```

### Task 3 (bonus)

There is an opinion that “they used to make a real movie, not like now.” Your task is to verify this statement using files with movie ratings from the last home lesson (the movies.csv and ratings.csv file from the grouplens.org.../movielens database). That is, to check whether it is true that with the growth of the year of release of the film, its average rating becomes lower.

At the same time, we will not touch on the subjective factors of setting these ratings, but will go through the following algorithm:

In the years variable, write a list of all the years from 1950 to 2010.

Write the production_year function, which sets the year of release for each line of the movie title. Not all movie titles contain the year of release in the same format, so use the following algorithm:

for each row, go through all the years in the years list
. if the year number is present in the movie title, the function returns this year as the year of release.
if none of the year numbers of the years list is found in the title of the film, then the year 1900 is returned.
Write down the year of release of the film according to the algorithm of paragraph 2 in the new column ‘year’

Calculate the average rating of all films for each value of the ‘year’ column and sort the result in descending order of rating

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')

df = ratings.merge(movies, on = 'movieId')
years = list(range(1950, 2011))

def production_year(title):
    for year in years:
        if str(year) in title:
            return year
    return 1900

df['year'] = df['title'].apply(production_year)

ratings_years = df[['rating', 'year']].groupby(by = 'year').mean().sort_values('rating', ascending=False)
ratings_years.head(10)
```
