### Task 1

There are 2 variables in which strings of arbitrary length are stored: phrase_1 and phrase_2.
Write code that checks which of these lines is longer.

Examples of how the program works:

phrase_1 = '640Kb should be enough for any tasks. Bill Gates (according to legend) '
phrase_2 = 'How much easier it would be to write programs if it weren't for customers'

Result: Phrase 2 is longer than phrase 1

```python
phrase_1 = 'Мне ль не знать,откуда сам я? Ненасытный, словно пламя'
phrase_2 = 'Сам собой охвачен весь. Свет есть всё, что я хватаю, Уголь всё, что отпускаю: Пламя - пламя, я и есмь!'

if len(phrase_1)>len(phrase_2):
  print('Фраза 1 длиннее фразы 2')
elif len(phrase_1)<len(phrase_2):
  print('Фраза 2 длиннее фразы 1')
else:
  print('Фразы равной длины')
```

Example of how the program works:

Leap year

```python
year = int(input('Укажите год '))

if year % 4 != 0:
    print('Не високосный год')
elif year % 100 != 0:
    print('Високосный год')
elif year % 400 == 0:
    print('Високосный год')
else:
    print('Не високосный год')
```

### Task 4

You need to write a program for selecting packages according to the size of the goods. Dimensions (width, length, height) are stored in variables (in centimeters):

Use the following rules:

if each of the three dimensions is less than or equal to 15 centimeters, then display "Box No. 1" on the screen;
if at least one of the measurements is more than 2 meters, then output "Ski packaging";
if at least one of the measurements is more than 15 centimeters, but less than 50 centimeters, then output "Box No. 2";
in all other cases, output "Box #3".
Example of how the program works:

width = 15
length = 55
height = 15
Result:
Standard box No. 3

width = 45
length = 205
height = 45
Result:
Looking for ski packaging

```python
while 1 == 1:
    
  width = int(input('width_cm '))
  length = int(input('length_cm '))
  height = int(input('height_cm '))
  size = [width, length, height]
  dins = max(size)
  if dins <= 15:
     print('Коробка №1')
  elif 15 < dins < 50:
     print('Коробка №2')
  elif dins > 200:
     print('Упаковка для лыж')
  else:
     print('Коробка №3')
  break
```

Examples of how the program works:

number = 123456
Result:
Unlucky ticket

number = 123321
Result:
Lucky ticket

```python
ticket_number = input('Введите номер билета (6 цифр): ')

number_1 = 0
number_2 = 0
left_number = 0
right_number = 0

if ticket_number.isdigit() and len(ticket_number) == 6:
  for number_1 in range(0,3):
    left_number = left_number + int(str(ticket_number)[number_1])  

  for number_2 in range(3,6):
    right_number = right_number + int(str(ticket_number)[number_2])

  if left_number == right_number:
    print('Счастливый билет')
  else:
    print('Несчастливый билет')
  
else:
  print('Неверный номер билета')
```

if a rectangle is the length of its two sides.
Example of how the program works:

Enter the shape type:
Triangle

2

2

3
Result:
The area of the triangle: 1.98

```python
import math
```

```python
tip = str(input('Круг, прямоугольник или треугольник? '))

if tip == 'треугольник':
    a = float(input('Введите точную длину стороны a = '))       
    b = float(input('Введите точную длину стороны b = '))       
    c = float(input('Введите точную длину стороны c = '))       
    p = (a+b+c)/2       
    s = math.sqrt((p*(p-a)*(p-b)*(p-c)))   
elif tip == 'прямоугольник':       
    a = float(input('Введите сторону a = '))       
    b = float(input('Введите сторону b = '))       
    s = a*b   
elif tip == 'круг':       
    r = float(input('Введите радиус r = '))       
    s = math.pi*(r**2)   

print(s)
```
